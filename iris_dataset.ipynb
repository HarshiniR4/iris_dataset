```python
#Exploratory data analytics of the iris flower dataset
#Objective: to be able to find out the class of the iris flower given a set of new data of a flower
#To identify the distinguishing features of each class of the iris flower

#Importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
os.chdir('D:\mini_task_1')
#after the setting the required working file read the csv file containing the iris flower dataset into a pandaas data frame
iris=pd.read_csv('iris_csv.csv')
```

```python
#Understanding the iris dataset
#to find the dimensions of the dataset and the number of data points 
iris.shape
```

```python
#to print the column headings of the dataset
iris.columns
```

```python
#displaying the first five data points along with the column headers of the datset
iris.head()
```

```python
#printing the basic information of the dataset
iris.info()
#from this we understand that there is no missing data in this datset
```

```python
#indentifying the types of classes of the iris flower and
#counting the datapoints in each of these classifications
iris["class"].value_counts()
#we see that all three classes of iris flowers have equal number of data points and hence this is a balanced dataset
```

```python
#to obtain the basic statistical information from the dataset
iris.describe()
```

```python
#seperating the table into three categories with accordance to the three classes of the iris flower
grouped=iris.groupby('class')
setosa=grouped.get_group("Iris-setosa")
versicolor=grouped.get_group("Iris-versicolor")
virginica=grouped.get_group("Iris-virginica")
#each of these dataframes contain the 50 datapoints of the flowers belonging to that class 
```

```python
#Understanding the three classes - setosa, virginica, versicolor
#since there are four features for every datapoint in this dataset,
#print pair plots to analyse how to differentiate the three classes
sns.set_style("whitegrid")
sns.pairplot(iris,hue="class",height=3)
plt.show()
#these plots help us differentiate iris-setosa from the other two classes on the basis of sepal length, sepal width, petal length and petal width 
```

```python
#the scatter plots are not very helpful in helping us differentiate the three classes of iris flowers

#to plot the histogram of sepal length of all datapoints in the data set
#histograms show the pdf-probablity density function of the datapoints 
sns.FacetGrid(iris,hue="class",height=3).map(sns.distplot,"sepallength").add_legend()
plt.show()
#we cannot obtain a clear distincion of the three classes from this plot
```

```python
#to plot the histogram of sepal width of all datapoints in the data set
sns.FacetGrid(iris,hue="class",height=3).map(sns.distplot,"sepalwidth").add_legend()
plt.show()
#this plots doesn't provide any clear difference in the sepal width characteristics of the three classes
```

```python
#to plot the histogram of the petal width of all the datapoints in the dataset
sns.FacetGrid(iris,hue="class",height=3).map(sns.distplot,"petalwidth").add_legend()
plt.show()
#in this plot we can distinguish iris-setosa petal width to a certain extent
```

```python
#to plot the histogram of petal length variations for all the datapoints in the dataset
sns.FacetGrid(iris,hue="class",height=3).map(sns.distplot,"petallength").add_legend()
plt.show()
#iris setosa can be clearly distinguished from the other two classes in accordance to their petallength 
#if the petal length of flower lies between 1cm and 2cm we can classify it under the class of iris-setosa
#hence, petal length is the better feature for distinguishing the three classes
```

```python
#to understand the classes better let us discuss the characteristics of the petal length of the three classes
#find the Means of each of the classes for their respective petal lengths 
print('Means:')
print(np.mean(setosa["petallength"]))
print(np.mean(versicolor["petallength"]))
print(np.mean(virginica["petallength"]))


#find the Standard Deviation from mean 
print('\nStandard Means:')
print(np.std(setosa["petallength"]))
print(np.std(versicolor["petallength"]))
print(np.std(virginica["petallength"]))

#find the Median 
print('\nMedians:')
print(np.median(setosa["petallength"]))
print(np.median(versicolor["petallength"]))
print(np.median(virginica["petallength"]))
```
